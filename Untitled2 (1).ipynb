```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
web=requests.get("https://www.magicbricks.com/property-for-sale/residential-real-estate?bedroom=&proptype=Multistorey-Apartment,Builder-Floor-Apartment,Penthouse,Studio-Apartment&cityName=Pune")
```

```python
web
```

```python
soup=BeautifulSoup(web.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
property_data=soup.find_all('div',class_="mb-srp__card__container")
```

```python
property_data
```

```python
property_name=[]
Detail=[]
```

```python
for p in property_data:
    name=p.find('h2',class_="mb-srp__card--title").text
    property_name.append(name)
```

```python
property_name
```

```python
for p in property_data:
    details=p.find('div',class_='mb-srp__card__summary__list summary-luxury').text
    Detail.append(details)
```

```python
Detail
```

```python
df_property=pd.DataFrame({"property Name":property_name,"Details":Detail})
```
